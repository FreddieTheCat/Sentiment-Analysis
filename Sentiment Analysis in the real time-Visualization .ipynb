```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook
plt.style.use('fivethirtyeight')
```

```python
frame_len = 10000

fig = plt.figure(figsize=(8,5))

def animate(i):
    data = pd.read_csv('List.csv')
    y1 = data['Positive']
    y2 = data['Negative']
    y3 = data['Neutral']  

    if len(y1)<=frame_len:
        plt.cla()
        plt.plot(y1, label='Positive')
        plt.plot(y2, label='Negative')
        plt.plot(y3, label='Neutral')
    else:
        plt.cla()
        plt.plot(y1[-frame_len: ], label='Positive')
        plt.plot(y2[-frame_len: ], label='Negative')
        plt.plot(y3[-frame_len: ], label='Neutral')
    
    plt.legend(loc='upper left')
    plt.tight_layout()


ani = FuncAnimation(plt.gcf(), animate, interval=500)
```
